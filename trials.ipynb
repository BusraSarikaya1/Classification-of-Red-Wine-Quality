```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('C:\Final_project\Final_project\data\winequality-red.csv')
```

For each variable in the dataset, it involves detecting outliers using the interquartile range (IQR) and then correcting these outliers according to the established rule.

```python
df_chlorides = df['chlorides']
plt.figure(figsize=(8,5))
plt.title('Aykırı Gözlemli')
sns.distplot(x = df_chlorides , color = 'blue')
Q1 = df_chlorides .quantile(0.25)
Q3 = df_chlorides .quantile(0.75)
IQR = Q3 - Q1
alt_sinir = Q1 - 1.5 * IQR
ust_sinir = Q3 + 1.5 * IQR
ust_aykiri = (df_chlorides  > ust_sinir)
ort = df_chlorides .mean()
medyan = df_chlorides .median()
df_chlorides [ust_aykiri] = medyan
df['chlorides'] = df_chlorides
plt.figure(figsize=(8,5))
plt.title('Aykırı Gözlemsiz')
sns.distplot(x = df_chlorides , color = 'blue')
```

# When this method is applied to "chlorides," the graph normalizes.
