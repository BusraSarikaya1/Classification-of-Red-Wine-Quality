```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
import plotly.express as px
import seaborn as sns
```

```python
df = pd.read_csv('C:\Final_project\Final_project\data\winequality-red.csv')
```

```python
data= df.copy()
```

```python
df = data.replace({'quality' : {
                                    8 : 'Quality',
                                    7 : 'Quality',
                                    6 : 'Quality',
                                    5 : 'Non-Quality',
                                    4 : 'Non-Quality',
                                    3 : 'Non-Quality',
        }}
)
```

```python
sns.boxplot(x=df['chlorides'],y=df['quality'])
```

```python
Q1 = df['chlorides'].quantile(0.25)
Q3 = df['chlorides'].quantile(0.75)
IQR = Q3 - Q1
low_cap = Q1 - 1.5 * IQR
upper_cap = Q3 + 1.5 * IQR
upper_outliers = (df['chlorides'] > upper_cap)
mean = df['chlorides'].mean()
median = df['chlorides'].median()
df['chlorides'][upper_outliers] = median
df['chlorides'] = df['chlorides']
```

- "chlorides" sütunundaki aykırı değerler, Tukey's Method kullanılarak tespit ediliyor ve yukarıdaki sınıra (upper_cap) çekiliyor. Yani, bu aykırı değerler, üst sınırdan büyük olan değerlere sahip olan gözlemler, bu sınıra çekilerek değiştiriliyor.
